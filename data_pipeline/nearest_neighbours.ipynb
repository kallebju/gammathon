```python
import os
import pandas as pd
```

```python
BASE_DIR = os.path.join(os.path.abspath(''), '..')
print(BASE_DIR)
```

```python
df_selected = pd.read_csv(BASE_DIR + '//data/guldhedenAnalytica.csv')
print(df_selected.head)
```

```python
df_customer = pd.read_csv(BASE_DIR + '//data/customer_data.csv')
print(df_customer)
```

```python
df_customer_last_month = df_customer.loc[df_customer.month == '2020-10-01']
df_customer_last_month
```

```python
df_customer_selected = df_customer_last_month[df_customer_last_month.customer_id.isin(df_selected.customer_id)]
df_customer_selected
```

```python
df_customer_not_selected = df_customer.drop(df_customer_selected.index)
df_customer_not_selected
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

for col in ['cluster']:
    df_customer_selected[col] = encoder.fit_transform(df_customer_selected[col])
    df_customer_not_selected[col] = encoder.transform(df_customer_not_selected[col])

```

```python
df_customer_not_selected = df_customer_not_selected[df_customer_not_selected['churned'] == False]
#df_customer_not_selected.drop(df_customer_not_selected['churned'] == True, axis =1, inplace=True)
df_customer_not_selected['churned'].describe()
```

```python
df_customer_selected.drop(['churned',], axis=1, inplace=True)
df_customer_not_selected.drop(['churned'], axis=1, inplace=True)

```

```python
df_customer_selected.drop(['month',], axis=1, inplace=True)
df_customer_not_selected.drop(['month'], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df_customer_selected[['months_since_joined', 'referred_friends', 'transaction_count', 'transaction_value', 'clv', 'total_reward_points', 'earned_reward_points']] = scaler.fit_transform(df_customer_selected[['months_since_joined', 'referred_friends', 'transaction_count', 'transaction_value', 'clv', 'total_reward_points', 'earned_reward_points']])
df_customer_not_selected[['months_since_joined', 'referred_friends', 'transaction_count', 'transaction_value', 'clv', 'total_reward_points', 'earned_reward_points']] = scaler.fit_transform(df_customer_not_selected[['months_since_joined', 'referred_friends', 'transaction_count', 'transaction_value', 'clv', 'total_reward_points', 'earned_reward_points']])

```

```python
df_customer_selected_A = df_customer_selected.loc[df_customer_selected.cluster == 0]
df_customer_selected_B = df_customer_selected.loc[df_customer_selected.cluster == 1]
df_customer_selected_C = df_customer_selected.loc[df_customer_selected.cluster == 2]

df_customer_not_selected_A = df_customer_not_selected.loc[df_customer_not_selected.cluster == 0]
df_customer_not_selected_B = df_customer_not_selected.loc[df_customer_not_selected.cluster == 1]
df_customer_not_selected_C = df_customer_not_selected.loc[df_customer_not_selected.cluster == 2]
```

```python
from sklearn.neighbors import NearestNeighbors
from scipy.spatial import KDTree
columns_to_compare = ['months_since_joined', 'referred_friends', 'transaction_count', 'transaction_value', 'clv', 'total_reward_points', 'earned_reward_points']


dataA_A = df_customer_selected_A[columns_to_compare]
dataB_A = df_customer_not_selected_A[columns_to_compare]

neigh = NearestNeighbors(n_neighbors=1)
neigh.fit(dataA_A) 
print(neigh.kneighbors(dataB_A)[1])
cluster_A_neighbours = neigh.kneighbors(dataB_A)[1]

dataA_B = df_customer_selected_B[columns_to_compare]
dataB_B = df_customer_not_selected_B[columns_to_compare]

neigh = NearestNeighbors(n_neighbors=1)
neigh.fit(dataA_B) 
print(neigh.kneighbors(dataB_B)[1])
cluster_B_neighbours = neigh.kneighbors(dataB_B)[1]

dataA_C = df_customer_selected_C[columns_to_compare]
dataB_C = df_customer_not_selected_C[columns_to_compare]

neigh = NearestNeighbors(n_neighbors=1)
neigh.fit(dataA_C) 
print(neigh.kneighbors(dataB_C)[1])
cluster_C_neighbours = neigh.kneighbors(dataB_C)[1]
```
